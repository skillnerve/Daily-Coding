Insertion Sort
    
    - Inplace Sort
    - Faster for less number of elements.

Screenshot (358).png

```python
q=[3,1,7,4,9,5]
for i in range(1,len(q)):
  w=q[i]
  j=i-1
  while j>=0 and q[j]>w:
    q[j+1]=q[j]
    j-=1
  q[j+1]=w
  print('STEP {}:{}'.format(i,q))
print(q)
```
